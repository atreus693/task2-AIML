```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Load dataset from seaborn
titanic = sns.load_dataset('titanic')

# Display first 5 rows
titanic.head()
```

```python
# Summary statistics for numerical features
print(titanic.describe())

# Summary for categorical features
print(titanic.describe(include='object'))
```

```python
numeric_features = ['age', 'fare']

# Histograms
for feature in numeric_features:
    plt.figure(figsize=(6,4))
    sns.histplot(titanic[feature].dropna(), kde=True)
    plt.title(f'Distribution of {feature}')
    plt.show()

# Boxplots
for feature in numeric_features:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=titanic[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

```python
# Pairplot with selected features
sns.pairplot(titanic[['age', 'fare', 'pclass', 'survived']].dropna(), hue='survived')
plt.show()
```

```python
# Survival rate by gender
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title('Survival Count by Gender')
plt.show()

# Survival by class
sns.countplot(x='pclass', hue='survived', data=titanic)
plt.title('Survival Count by Passenger Class')
plt.show()

# Fare vs Survival (Plotly for interactivity)
fig = px.box(titanic, x="survived", y="fare", color="survived", 
             labels={'survived': 'Survived', 'fare': 'Fare'}, title="Fare by Survival")
fig.show()
```
